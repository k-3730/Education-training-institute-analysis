```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://teacher.i-scream.co.kr/help/afterword/list.do?sso=ok")
driver.implicitly_wait(30) # 30초 
```

```python
reviews = []
page_num = 3

# 페이지 넘기기
for j in range(1, 13):
    for i in range(1, 11):  # 페이지 크롤링

        # 페이지 링크를 직접 클릭
        new_tab = driver.find_element(by='xpath', value=f"""//*[@id="content"]/div[3]/table/tbody/tr[{i}]/td[3]/a""")
        new_tab.click()

        WebDriverWait(driver, 25).until(EC.presence_of_element_located((By.XPATH, """//*[@id="bbs_wrap"]/div/dl[2]/dd/a""")))

        title = driver.find_element(by='xpath', value="""//*[@id="bbs_wrap"]/div/dl[2]/dd/a""").text
        element = driver.find_element(by='xpath', value="""//*[@id='bbs_wrap']/div/dl[3]/dd[1]/span/span""")
        
        # 점수 추출 로직
        class_attr = element.get_attribute("class")
        grade = int(class_attr.split("rate")[-1])
        
        content = driver.find_element(by='xpath', value="""//*[@id="bbs_wrap"]/div/div[1]""").text
        reviews.append([title, grade, content])

        # 목록으로 돌아가기 위한 버튼 
        back_to_list_button = driver.find_element(by='xpath', value="""//*[@id="btn-list"]""")
        back_to_list_button.click()

        WebDriverWait(driver, 35).until(EC.presence_of_element_located((By.XPATH, f"""//*[@id="paging"]/ul/li[{page_num}]/a""")))

    # 페이지 번호가 2일 경우만 Xpath가 다르므로 별도 처리
    if page_num == 2:  
        next_button = driver.find_element(by='xpath', value="""//*[@id="paging"]/ul/li[2]/span""")
    else:
        next_button = driver.find_element(by='xpath', value=f"""//*[@id="paging"]/ul/li[{page_num}]/a""")
    next_button.click()

    if j % 10 == 0:  # 10페이지, 20페이지, 30페이지 등의 경우
        # '다음' 버튼을 클릭
        next_button = driver.find_element(by='xpath', value="""//*[@id="paging"]/ul/li[12]/a""")
        next_button.click()
        page_num = 2  # 다음 버튼을 누른 후 페이지 번호를 2로 초기화
    else:
        page_num += 1

    if page_num > 11:  # 일반적인 경우, 페이지 번호가 11을 초과하면 3으로 초기화
        page_num = 3
```
