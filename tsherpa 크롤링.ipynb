```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://edu.tsherpa.co.kr/LectureInfo/LectureReviews2")
driver.implicitly_wait(30) # 30초 기다리기
```

동적대기 미적용

```python
reviews = []
page_num = 3

# 페이지 넘기기
for j in range(1,31):
    for i in range(1,11): # 페이지 크롤링
        main_window_handle = driver.current_window_handle
        
        new_tab = driver.find_element(
        by = 'xpath',
        value = f"""//*[@id="mainForm"]/table/tbody/tr{[i]}/td[2]/a""")

        new_tab.click()

        time.sleep(5)

        for handle in driver.window_handles:
            if handle != main_window_handle:
                new_window_handle = handle
                break
        
        driver.switch_to.window(new_window_handle)

        title = driver.find_element(
            by = 'xpath',
            value = f"""//*[@id="mainForm"]/div[2]/table/tbody/tr[1]/td""").text
        
        grade_element = driver.find_element(by='xpath', value='//*[@id="mainForm"]/div[2]/table/tbody/tr[2]/td/span[1]/img')
        grade = grade_element.get_attribute('alt')
        
        content = driver.find_element(
            by = 'xpath',
            value = f"""//*[@id="mainForm"]/div[2]/table/tbody/tr[3]/td""").text

        reviews.append([title,grade,content])

        driver.close()

        driver.switch_to.window(main_window_handle)

    next_button = driver.find_element(
        by='xpath', 
        value=f"""//*[@id="mainForm"]/div[3]/a{[page_num]}""")
    
    next_button.click()
    
    time.sleep(5)

    if page_num < 12:
        page_num += 1
    else:
        page_num = 3 
```

```python
pd.DataFrame(reviews)
```

```python
review_01_30 = pd.DataFrame(reviews)
```

```python
review_01_30.to_excel(excel_writer='01-30.xlsx')
```

동적대기 적용

```python
reviews = []
page_num = 3

# 페이지 넘기기
for j in range(1,20):
    for i in range(1,11): # 페이지 크롤링
        main_window_handle = driver.current_window_handle

        new_tab = WebDriverWait(driver, 25).until(
            EC.presence_of_element_located((By.XPATH, f"""//*[@id="mainForm"]/table/tbody/tr[{i}]/td[2]/a"""))
        )
        new_tab.click()

        WebDriverWait(driver, 20).until(EC.number_of_windows_to_be(2))

        for handle in driver.window_handles:
            if handle != main_window_handle:
                new_window_handle = handle
                break
        
        driver.switch_to.window(new_window_handle)

        title = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, f"""//*[@id="mainForm"]/div[2]/table/tbody/tr[1]/td"""))
        ).text

        grade_element = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="mainForm"]/div[2]/table/tbody/tr[2]/td/span[1]/img'))
        )
        grade = grade_element.get_attribute('alt')
        
        content = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, f"""//*[@id="mainForm"]/div[2]/table/tbody/tr[3]/td"""))
        ).text

        reviews.append([title,grade,content])

        driver.close()

        driver.switch_to.window(main_window_handle)

    next_button = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, f"""//*[@id="mainForm"]/div[3]/a[{page_num}]"""))
    )
    next_button.click()

    if page_num < 12:
        page_num += 1
    else:
        page_num = 3 
```

```python
review_271_289 = pd.DataFrame(reviews)
```

```python
review_271_289.to_excel(excel_writer='271-289.xlsx')
```
