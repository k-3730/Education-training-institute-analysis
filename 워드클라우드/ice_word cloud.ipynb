```python
text = ' '.join([word for word in text.split() if len(word) >= 8])

# 워드클라우드 설정 및 생성
font_path = "C:/Windows/Fonts/malgun.ttf"
wordcloud = WordCloud(
    width=1200, 
    height=800, 
    background_color="white", 
    max_words=100, 
    font_path=font_path
).generate(text)

# 워드클라우드 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
