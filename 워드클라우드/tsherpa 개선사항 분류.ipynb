```python
import pandas as pd
import re
```

```python


def is_improvement_request(text):
    negative_pattern = r"(제가).*?어(려운|려워|렵)|(제가).*?(부족했)|반성"
    if re.search(negative_pattern, text):
        return False
    
    pattern = r"(그래도).*?(괜찮|나쁘)|하지만|아(쉬웠|쉬었|쉬운|쉽)|(부족|필요).*?(같아)|(좋|있)으나|좋았(는데|지만)|개선|불편|다만|비추천|실망|지루(한|했)|수정(이|되)|별로(였|네)"
    
    # 텍스트에서 패턴 검색
    if re.search(pattern, text):
        return True
    return False



# 엑셀 파일 불러오기
df = pd.read_excel("천재교육 연수원 후기최종.xlsx")

# 후기 컬럼에서 개선 요청이 포함된 행들만 필터링
improvement_reviews = df[df["후기"].apply(is_improvement_request)]
```

```python
# 데이터 프레임 생성
result_df = pd.DataFrame({
    'Category': ['개선 요청 포함 후기'],
    'Count': [improvement_reviews]
})

pd.DataFrame(improvement_reviews)
```

```python

# 필요하다면 개선 요청 포함된 후기들만 새로운 엑셀 파일로 저장
improvement_reviews.to_excel("개선요청_후기.xlsx", index=False)
```
